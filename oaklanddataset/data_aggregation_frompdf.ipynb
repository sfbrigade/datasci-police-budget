```python
import pandas as pd
import numpy as np
```

```python
oakland = pd.read_csv('../data/processed/oakland-budget_frompdf.csv')
```

```python
oakland
```

```python

#Create the Public Protection Category
oakland.loc[(oakland['Department'] == 'Race & Equity') | 
            (oakland['Department'] == 'Department of Violence Prevention') |
            (oakland['Department'] == 'Fire Department') |
            (oakland['Department'] == 'Police Commission') |
            (oakland['Department'] == 'Police Department'), 'Agencies by Category'] = 'Public Protection'


#Create the Public Works, Transportation, and Commerce
oakland.loc[(oakland['Department'] == 'Capital Improvement Projects') | 
            (oakland['Department'] == 'Transportation') |
            (oakland['Department'] == 'Economic & Workforce Development') |
            (oakland['Department'] == 'Non-Departmental') |
            (oakland['Department'] == 'Oakland Public Works'), 'Agencies by Category'] = 'Public Works, Transportation, and Commerce'
    
    
#Create the Human Welfare & Neighborhood Development
oakland.loc[(oakland['Department'] == 'Housing & Community Development') | 
            (oakland['Department'] == 'Human Services'), #called human services instead of human rights commission
             'Agencies by Category'] = 'Human Welfare & Neighborhood Development'

#Create Culture & Recreation
oakland.loc[(oakland['Department'] == 'Oakland Public Library') | 
            (oakland['Department'] == 'Oakland Parks, Recreation & Youth Development'), 
            'Agencies by Category'] = 'Culture & Recreation'
    
        
#Create General Administration & Finance
oakland.loc[(oakland['Department'] == 'City Council') | 
            (oakland['Department'] == 'City Administrator') |
            (oakland['Department'] == 'City Attorney') |
            (oakland['Department'] == 'City Auditor') |
            (oakland['Department'] == 'City Clerk') |
            (oakland['Department'] == 'Planning & Building') |
            (oakland['Department'] == 'Public Ethics Commission') |
            (oakland['Department'] == 'Finance Department') |
            (oakland['Department'] == 'Human Resources Management Department') |
            (oakland['Department'] == 'Information Technology') |
            (oakland['Department'] == 'Mayor'),
            'Agencies by Category'] = 'General Administration & Finance'


    
    
    
    
    
    
    
    
    
    
    
```

```python
oakland
```

```python
oakland.columns
```

```python
agencies = oakland.groupby(['Agencies by Category'])[['FY2017-18 ', 'FY2018-19', 'FY2019-20', 'FY2020-21']].agg('sum')
```
