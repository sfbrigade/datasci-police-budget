### Now we will look at Public Protection more closely. 
These five divisions are:
1. Race & Equity
2. Department of Violence Prevention
3. Fire Department
4. Police Comission
5. Police Department

#### Background Information on Public Protection

##### The current working total(FY20-21) of Public protection is 540 million. 

This division of the sector emphasizes that the vast majority of resources are allocated to the Police Department especially when compared to Race & Equity and the Department of Violence Prevention. 


The Police Commision oversees the Police Department. This includes the Office of the Inspector General, which audits the police department as well as reviews policies/procedures. This is where citizens go to file police misconduct reports. This sector was given $4,320,062 in 2020-2021.


There is a different department called the Department of Violence Prevention. The goals of this department are to reduce homicides, gun related violence, domestic violence, and exploitation of children. This department was given $1,177,495 in 2020-2021. 

The field operations of the fire department are involved in fire suppression, emergency medical response, and rescue. This department was given $203,370,848 as part of the biennial working total in 2020-2021.

 Race & Equity has consistently had the lowest budget in the available dataset. This department was given $759,313 as part of the biennial working total in 2020-2021.

```python
#Create smaller dataframe with percentages added for each year

#2018-2019
fy1819 = oakland[['Department Alias', 'FY18-19 Midcycle Adopted Total', 'Agencies by Category']].copy()
fy1819['18-19 Percent'] = fy1819['FY18-19 Midcycle Adopted Total']/fy1819['FY18-19 Midcycle Adopted Total'].sum() * 100

#2019-2020
fy1920 = oakland[['Department Alias', 'FY19-20 Biennial Working Total', 'Agencies by Category']].copy()
fy1920['19-20 Percent'] = fy1920['FY19-20 Biennial Working Total']/fy1920['FY19-20 Biennial Working Total'].sum() * 100

#2020-2021
fy2021 = oakland[['Department Alias', 'FY20-21 Biennial Working Total', 'Agencies by Category']].copy()
fy2021['20-21 Percent'] = fy2021['FY20-21 Biennial Working Total']/fy2021['FY20-21 Biennial Working Total'].sum() * 100
```

```python
fy1819 = fy1819[fy1819['Agencies by Category'] == 'Public Protection'].drop(columns='Agencies by Category')
fy1920 = fy1920[fy1920['Agencies by Category'] == 'Public Protection'].drop(columns='Agencies by Category')
fy2021 = fy2021[fy2021['Agencies by Category'] == 'Public Protection'].drop(columns='Agencies by Category')
```

### Each of the percentage divisions as dataframes

```python
#Fix the indexes
fy1920.index = fy1920['Department Alias']
fy1819.index = fy1920['Department Alias']
fy2021.index = fy1920['Department Alias']

#Create new dataframe for visualizations

df_excl = pd.DataFrame()
df_excl = df_excl.append(fy1819['18-19 Percent'])
df_excl = df_excl.append(fy1920['19-20 Percent'])
df_excl = df_excl.append(fy2021['20-21 Percent'])

#create a new column called other to compare the divisions to

df_excl['other'] = 100 - df_excl.sum(axis=1)
df_excl
```

```python
### Data Visualizations of of the percentages compared to the total

df_transp = df_excl.T

df_transp

plt.rcParams.update({'font.size': 10})
2017
amt = df_transp['19-20 Percent']
labels = df_transp.index
fig, ax1 = plt.subplots()
ax1.pie(amt, labels=labels, shadow=True, autopct='%1.1f%%')
ax1.axis('equal')
plt.title('2018-2019 Adopted Total Budgets')
plt.show()


# 2016

amt = df_transp['20-21 Percent']
labels = df_transp.index
fig, ax1 = plt.subplots()
ax1.pie(amt, labels=labels, shadow=True, autopct='%1.1f%%')
ax1.axis('equal')
plt.title('2018-2019 Adopted Total Budgets')
plt.show()


# 2015

amt = df_transp['18-19 Percent']
labels = df_transp.index
fig, ax1 = plt.subplots()
ax1.pie(amt, labels=labels, shadow=True, autopct='%1.1f%%')
ax1.axis('equal')
plt.title('2018-2019 Adopted Total Budgets')
plt.show()
```

```python
subset = oakland[oakland['Agencies by Category'] == 'Public Protection'] 
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.gca() 

ax.set_title(label='Public Protection Divisions') # Title

subset.plot(x='Department Alias', 
             y=['FY18-19 Midcycle Adopted Total', 
                'FY19-20 Biennial Working Total', 
                'FY20-21 Biennial Working Total'], 
             kind='bar',
            figsize=(15,10),
            rot=True,
            ax=ax,
           title='Public Protection Divisions')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
subset.index = subset['Department Alias']
#remove the agencies by category column
subset = subset.drop('Agencies by Category', axis=1)
```

```python
subset
```

```python
#transpose the dataframe
df_transp = subset.T

#put years in its own column
df_transp['Year'] = df_transp.index
#remove duplicated row
df_transp = df_transp.drop('Department Alias')
```

```python
df_transp
```

```python
years = df_transp['Year']

plt.figure(figsize=(20,15))
plt.rcParams.update({'font.size': 22})
plt.title('Public Protection', fontsize=40)
plt.xlabel('Year', fontsize=40)
plt.ylabel('Budget', fontsize = 40)
plt.tick_params(axis='both', which='major', labelsize=30)

vio = 'Department of Violence Prevention'
plt.plot(years, np.array(df_transp[vio]),
         label=vio)

fire = 'Fire Department'
plt.plot(years, np.array(df_transp[fire]),
         label=fire)

pol_c='Police Commission'
plt.plot(years, np.array(df_transp[pol_c]), 
         label=pol_c)

pol_d='Police Department'
plt.plot(years, np.array(df_transp[pol_d]), 
         label=pol_d)

race='Race & Equity'
plt.plot(years, np.array(df_transp[race]),
         label=race)
plt.legend(prop={'size': 20})
plt.show()
```
