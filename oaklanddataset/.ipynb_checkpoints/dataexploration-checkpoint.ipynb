The purpose of this notebook is to inspect and aggregate [data](https://data.oaklandnet.com/Financial/FY-2019-21-Biennial-Proposed-Budget/aeya-c8st) concerning the budget in Oakland, California.

### Import Modules and read the csv

```python
import pandas as pd
df = pd.read_csv('FY_2019-21_Adopted_Budget.csv')
```

```python
df.info()
```

```python
df.head(5)
```

#### Breakdowns of various departments that the budget goes towards

```python
departments = df['Department Alias'].unique()
```

```python
departments
```

```python
grouped = df.groupby('Department Alias')
```

### Inspect the various subdivisions of 'Police Department' and 'Police Commission'

Inside of this data set there are two different subdivisions of Police data. These are either in the Department Alias as 'Police Department' or 'Police Commision'. It is important to inspect these further to determine further subdivisions. 

```python
police_deparment = grouped.get_group('Police Department')
```

Overview of police department

```python
police_deparment.head()
```

Count unique values in 'Fund and Desc' in descending order

```python
police_deparment['Fund and Desc'].value_counts()
```

Count unique values in 'Division' column in descending order

```python
police_deparment['Division'].value_counts()
```

### Inspect Police Commision Category

```python
police_commission  = grouped.get_group('Police Commission')

police_commission.head()
```

Count unique values in 'Fund and Desc' in descending order

```python
police_commission['Fund and Desc'].value_counts()
```

Count unique cells in 'Org and Desc'

```python
police_commission['Org and Desc'].value_counts()
```

Count unqiue cells in 'Service Area Alias

```python
police_commission['Service Area Alias']
```

```python
aggregated_data = df.groupby(['Department Alias'])[['FY18-19 \nMidcycle Adopted Ongoing', 'FY18-19 \nMidcycle Adopted One Time', 'FY18-19 \nMidcycle Adopted Total', 'FY19-20 \nBiennial Working Ongoing','FY19-20 \nBiennial Working One Time','FY19-20 \nBiennial Working Total','FY20-21 \nBiennial Working Ongoing','FY20-21 \nBiennial Working One Time', 'FY20-21 \nBiennial Working Total']].agg('sum')
```

```python
aggregated_data
```

#### Further aggregation of the data to look only at the total budgets between 2018 and 2021

```python
final_budget = df.groupby(['Department Alias'])[['FY18-19 \nMidcycle Adopted Total', 
                                                 'FY19-20 \nBiennial Working Total',
                                                 'FY20-21 \nBiennial Working Total']].agg('sum')







```

```python
final_budget
```

### Export the Aggregated data into a new csv. 

In a separate jupyter notebook there will be visualizations to present to the Code America team.
