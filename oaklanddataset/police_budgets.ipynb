### This notebook will look specifically at police budgets and the divisions they are made of.

```python
import pandas as pd
df = pd.read_csv('FY_2019-21_Adopted_Budget.csv')
```

```python
df.info()
```

```python
df.head(10)
```

```python
df['Department Alias'].unique()
```

```python
df['Department Alias'].value_counts()
```

This is the aggregation into a police budget dataframe.

```python
pol_budget = df[(df['Department Alias'] == 'Police Department') | (df['Department Alias'] == 'Police Commision')]
```

```python
pol_budget
```

```python
pol_budget['Org and Desc'].value_counts()
```

```python
pol_budget['Service Area Alias'].value_counts()
```

```python
police_breakdown = pol_budget.groupby(['Org and Desc'])[['FY18-19 \nMidcycle Adopted Total', 
                                                 'FY19-20 \nBiennial Working Total',
                                                 'FY20-21 \nBiennial Working Total']].agg('sum')







```

```python
police_breakdown
```

```python
police_breakdown.rename(columns = {'FY18-19 \nMidcycle Adopted Total':'FY18-19 Midcycle Adopted Total', 
                          'FY19-20 \nBiennial Working Total':'FY19-20 Biennial Working Total', 
                          'FY20-21 \nBiennial Working Total':'FY20-21 Biennial Working Total'}, inplace = True) 
```

```python
police_breakdown.to_csv('oakland_pol_budget_breakdown.csv')
```

```python
df['Department'].unique()
```

```python
police_breakdown
```

```python
police_breakdown = police_breakdown.reset_index()
```

```python
import matplotlib.pyplot as plt

police_breakdown.plot(x='Org and Desc', 
             y=['FY18-19 Midcycle Adopted Total', 
                'FY19-20 Biennial Working Total', 
                'FY20-21 Biennial Working Total'], 
             kind='barh',
            figsize=(9,9))
plt.show()
```
