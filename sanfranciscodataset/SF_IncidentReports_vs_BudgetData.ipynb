```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style 
style.use('ggplot')
import seaborn as sns
```

```python
df_budget = pd.read_csv('../data/Spending_Revenue_cleaned.csv')
df_budget.drop(columns=['Unnamed: 0'], inplace=True)
df_budget.head()
```

```python
df_pol = df_budget[df_budget['Department'] == 'POL Police']
df_pol
```

```python
for i in range(len(years)):
    year = int(years[i])
    temp = year_dfs[i]['category']
    year_dfs[i]['category'] = year_dfs[i].index
    year_dfs[i][year] = temp
    year_dfs[i] = year_dfs[i].reset_index(drop=True)
```

```python
cat = year_dfs[0]['category']
```

```python
df_i_by_y = pd.concat(year_dfs, axis=1)
df_i_by_y
```

```python
df_i_by_y.drop(['category'], axis=1, inplace=True)
df_i_by_y['category'] = cat
```
