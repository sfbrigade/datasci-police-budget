```python
years = df['Fiscal Year'].unique()
years
```

```python
df_aggregated_sums = df.groupby(['Department'])[['Amount']].agg('sum')
df_aggregated_sums.columns = ['Amount_Sum_199-2017']
df_aggregated_sums
```

These sums are the sums of each departments amount from 1999 - 2017

```python
police_dep['Amount'].sum()
```

```python
df_year = []
for index, year in enumerate(years):
    df_year.append( df[df['Fiscal Year'] == year])
```

```python
df_year[0]
```

```python
len(df_year)
```

```python
department_sums = []
for i in range(19):
    department_sums.append( pd.DataFrame(df_year[i].groupby('Department').Amount.agg('sum')) )
    col_year = years[i]
    department_sums[i].columns = [col_year]
```

```python
df_aggregated=pd.concat(department_sums, axis=1)
df_aggregated
```

```python
department_sums[0]
```

It looks like there are departments that did not hav any funding in previous years. To solve this, I replaced the NaN values with 0

```python
df_aggregated = df_aggregated.fillna(0)
```

```python
df_aggregated['Department'] = df_aggregated.index
```

```python
df_aggregated.head()
```

```python
cols = df_aggregated.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_aggregated = df_aggregated[cols]
```
