```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('ggplot')
```

```python
shootings = pd.read_csv('../data/OfficerInvolvedShootings.csv')
shootings
```

```python
shootings = shootings.sort_values(by='Year')
```

```python
df_pop = pd.read_csv('sf_population_count.csv')
df_pop.drop(columns=['Unnamed: 0'], inplace=True)
df_pop
```

```python

plt.title('Population vs Officer Involved Shootings')
plt.plot(df_pop['year'], df_pop['population'])
plt.plot(shootings['Year'], shootings['Total'])
```

```python
# create figure and axis objects with subplots()
fig, ax = plt.subplots()
# make a plot
ax.plot(shootings['Year'], shootings['Total'], color="red", marker="o")
# set x-axis label
ax.set_xlabel("Year",fontsize=14)
# set y-axis label
ax.set_ylabel("Population Count in SF",color="red",fontsize=14)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(df_pop['year'], df_pop['population'],color="blue",marker="o")
ax2.set_ylabel("Number of Incident Reports",color="blue",fontsize=14)

plt.title('SF_Police_Shootings_vs_Population')
plt.savefig("../images/SF_Police_Shootings_vs_Population.png")
plt.show()
```

```python
plt.title('Population / Officer Involved Shootings')
pop_v_shootings = np.divide(df_pop['population'], shootings['Total'])
plt.plot(df_pop['year'], pop_v_shootings[:10])
plt.savefig("../images/Population_Divided_SF_Police_Shootings_.png")
plt.show()
```
