```python
yearly_analysis = unique_incidents.groupby(unique_incidents.Date.dt.year)
```

```python
import matplotlib.pyplot as plt
```

```python
crime_yearly_analysis = data.groupby([data.Date.dt.year, data.Category])
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.set_ylim(bottom=0)
ax.set_ylim(top=60000)

crime_yearly_analysis.count()["PdId"].unstack().plot(ax=ax, kind='bar')
```

```python
top_10_cats = data.Category.value_counts()[0:9].index.tolist()
```

```python
top_10_cats
```

```python
top_10_cat_incidents = data[data["Category"].isin(top_10_cats)]
```

```python
top_10_yearly = top_10_cat_incidents.groupby([top_10_cat_incidents.Date.dt.year, top_10_cat_incidents.Category])
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.set_ylim(bottom=0)
ax.set_ylim(top=60000)

top_10_yearly.count()["PdId"].unstack().plot(ax=ax, kind='bar')
```
